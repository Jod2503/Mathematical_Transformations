```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('concrete_data.csv')
df.head()
```

```python
# perform linear regression and check accuracy before transform
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:8],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train.isnull().sum()
```

```python
y_train.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# Box-Cox and Yeo johnson transform
```

```python
# by default, yeo johnson is applied
# this class automatically performs scaling, so no seperate need
```

```python
# check if any col is zero anywhere
```

```python
(df==0).sum()
```

```python
# hence as n>0 add a very small number 0.000001 to make >0
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='box-cox')
```

```python
x_train_transformed= pt.fit_transform(x_train+0.0000001)
x_test_transformed= pt.transform(x_test+0.0000001)
```

```python
pd.DataFrame({'cols':x_train.columns,'lambdas':pt.lambdas_})
```

```python
# apply regression
lr.fit(x_train_transformed,y_train)
y_pred=lr.predict(x_test_transformed)
```

```python
r2_score(y_test,y_pred)
```
